## Step 1: Importing the required libraries

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

## Step 2: Loading and exploring the data

```python
#pandas UnicodeDecodeError: 'utf-8' codec can't decode byte 0x97 in position 6785: invalid start byte.
#The error might have several different reasons: different encoding. bad symbols. corrupted file. Unicode escape finds the quotes in cells
#and remove them 
df=pd.read_csv("data.csv",encoding= 'unicode_escape')
```

## Step 3: Cleaning the Data

```python
# Stripping extra spaces in the description
df['Description'] = df['Description'].str.strip()
```

```python
df.isnull().sum()
```

```python
# Dropping the rows without any invoice number
df.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
```

```python
df['InvoiceNo']=df['InvoiceNo'].astype('str')
```

```python
# Dropping all transactions which were done on credit
df=df[~df['InvoiceNo'].str.contains('C')]
```

```python
# Default value of display.max_rows is 10 so at max
# 10 rows will be printed. Set it None to display
# all rows in the dataframe
pd.set_option('display.max_rows', None)
```

## Step 4: Splitting the data according to the region of transaction

```python
# Transactions done in France
#Unstack() convert the second column rows to column names
basket_France=(df[df['Country']=='France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().fillna(0).reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket_France
```

```python
# Transactions done in Portugal
basket_Por = (df[df['Country'] =="Portugal"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket_Sweden = (df[df['Country'] =="Sweden"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

## Step 5: Hot encoding the Data

```python
# Defining the hot encoding function to make the data suitable 
# for the concerned libraries
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
  
# Encoding the datasets
#DataFrame.applymap() is another approach to update all DataFrame columns at a time
basket_encoded = basket_France.applymap(hot_encode)
basket_France = basket_encoded
  
basket_encoded = basket_Por.applymap(hot_encode)
basket_Por = basket_encoded
  
basket_encoded = basket_Sweden.applymap(hot_encode)
basket_Sweden = basket_encoded
```

```python
basket_France
```

## Step 6: Building the models and analyzing the results

### a) France:

```python
# Building the model
frq_items=apriori(basket_France,min_support=0.05,use_colnames=True)
```

```python
frq_items
```

```python
# Collecting the inferred rules in a dataframe
#The IF component of an association rule is known as the antecedent. The THEN component is known as the consequent. 
#The antecedent and the consequent are disjoint; they have no items in common.
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
```

```python
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
```

```python
rules
```

#### From the above output, Paper cups, paper plates are brought together and both are related. The main reason for this is French governemnt banned the plastic in country.

### b) Portugal

```python
frq_items = apriori(basket_Por, min_support = 0.05, use_colnames = True)
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```

#### From the above output the Knick knack tins and color pencils are related.These belongs to primary school kids. Thus, logically make sense to be paired together.

### c.) Sweden

```python
frq_items = apriori(basket_Sweden, min_support = 0.05, use_colnames = True)
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```

#### From the above output, it depicts that boys and girls cutlery are paired together. This makes sense when parents goes shopping for cutlery for their children, they want the product to be little customized as per the kids wishes
